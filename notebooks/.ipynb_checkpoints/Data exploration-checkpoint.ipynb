```python
xmax0, xmax1, ymax0, ymax1 = 4, 4, 5, 5
x0, x1, y0, y1 = 0, x0+xmax0+1, 0, 0
figWidth = x0 + xmax0 + xmax1
figHeight = 5
fig = plt.figure(figsize=(figWidth,figHeight))
x0/= figWidth
x1/=figWidth
y0/= figHeight
y1/= figHeight
xmax0/= figWidth
xmax1/= figWidth
ymax0/= figHeight
ymax1/= figHeight
ax1 = plt.axes([x0, y0, xmax0, ymax0])
ax2 = plt.axes([x1, y1, xmax1, ymax1])
ax1.hist(data[' Age  (years) '].values, bins=20, color = 'lightslategray')
ax1.set_xlabel('Age (years)', fontproperties = fontlarge)
ax1.set_ylabel('Number of patients', fontproperties = fontlarge)
ax1.set_title('Age (distribution)', fontproperties = fontlarge)
ax2.bar(['Male', 'Female'], [len(data[data['Gender']=='M'].values), len(data[data['Gender']=='F'].values)], width=0.4, 
        color='skyblue')
ax2.set_ylabel('Number of patients', fontproperties = fontlarge)
ax2.set_title('Gender (distribution)')
```

```python
def identifyPatient(x):
    if 'PD' in x:
        return 'PD'
    elif 'RBD' in x:
        return 'RBD'
    else:
        return 'HC'
data['patientType'] = data[' Participant  code '].apply(identifyPatient)
```

```python
excludecols = [' Participant  code ', 'patientType']
includecols = [i for i in data.columns if i not in excludecols]
categorical = ['Gender',' Positive  history  of  Parkinson  disease  in  family ',
       ' Antidepressant  therapy ', ' Antiparkinsonian  medication ',
       ' Antipsychotic  medication ', ' Benzodiazepine  medication ']
```

```python
results = pd.DataFrame()
i=0
for col in includecols:
    if col not in categorical:
        data[col] = data[col].apply(lambda x: float(x) if x!='-' else np.nan)
        results.loc[i, 'ColName'] = col
        results.loc[i, 'Parkinson Disease'] = data[data['patientType']=='PD'][col].mean()
        results.loc[i, 'RBD'] = data[data['patientType']=='RBD'][col].mean()
        results.loc[i, 'Health Controls'] = data[data['patientType']=='HC'][col].mean()
        if any(pd.isnull(data[data['patientType']=='HC'][col].values)):
            if any(pd.isnull(data[data['patientType']=='RBD'][col].values)):
                results.loc[i, 'pvalue_PD_RBD'] = np.nan
                results.loc[i, 'pvalue_PD_HC'] = np.nan
                results.loc[i, 'pvalue_RBD_HC'] = np.nan
            else:
                results.loc[i, 'pvalue_PD_RBD'] = ttest_ind(data[data['patientType']=='PD'][col].values, data[data['patientType']=='RBD'][col].values)[1]
                results.loc[i, 'pvalue_PD_HC'] = np.nan
                results.loc[i, 'pvalue_RBD_HC'] = np.nan
        else:
            results.loc[i, 'pvalue_PD_RBD'] = ttest_ind(data[data['patientType']=='PD'][col].values, data[data['patientType']=='RBD'][col].values)[1]
            results.loc[i, 'pvalue_PD_HC'] = ttest_ind(data[data['patientType']=='PD'][col].values, data[data['patientType']=='HC'][col].values)[1]
            results.loc[i, 'pvalue_RBD_HC'] = ttest_ind(data[data['patientType']=='RBD'][col].values, data[data['patientType']=='HC'][col].values)[1]
        i+=1
    else:
        pdp=[]
        rbd=[]
        hc=[]
        for cat in data[col].unique():
            results.loc[i, 'ColName'] = f'{col}_{cat}'
            results.loc[i, 'Parkinson Disease'] = len(data[(data['patientType']=='PD') & (data[col]==cat)])
            results.loc[i, 'RBD'] = len(data[(data['patientType']=='RBD') & (data[col]==cat)])
            results.loc[i, 'Health Controls'] = len(data[(data['patientType']=='HC') & (data[col]==cat)])
            pdp.append(results.loc[i, 'Parkinson Disease'])
            rbd.append(results.loc[i, 'RBD'])
            hc.append(results.loc[i, 'Health Controls'])
            i+=1
        try:
            results.loc[i, 'pvalue_PD_RBD'] = chi2_contingency([pdp, rbd])[1]
            results.loc[i, 'pvalue_PD_HC'] = chi2_contingency([pdp, hc])[1]
            results.loc[i, 'pvalue_RBD_HC'] = chi2_contingency([hc, rbd])[1]
        except:
            results.loc[i, 'pvalue_PD_RBD'] = np.nan
            results.loc[i, 'pvalue_PD_HC'] = np.nan
            results.loc[i, 'pvalue_RBD_HC'] = np.nan
        
```
